```python
def parse_list_string(value):
    """
    Convert string representation of Python objects back to real objects.
    """
    if pd.isna(value):
        return []

    if isinstance(value, list):
        return value

    return ast.literal_eval(value)

df["skills_normalized"] = df["skills_normalized"].apply(parse_list_string)
```

## Skill Level Score
Skill levels are converted into numerical values:
- Basic = 1
- Intermediate = 2
- Advanced = 3
The total score represents the overall technical skill requirement of a job.

```python
skill_level_mapping = {
    "Basic": 1,
    "Intermediate": 2,
    "Advanced": 3
}

def calculate_skill_level_score(skills):
    """
    Calculate total skill level score for a job.
    """
    score = 0
    for skill in skills:
        level = skill.get("level")
        score += skill_level_mapping.get(level, 0)
    return score

df["skill_level_score"] = df["skills_normalized"].apply(calculate_skill_level_score)
```

```python
df["skill_level_score"].describe()
```

## Advanced Skill Ratio
This feature measures the proportion of advanced-level skills required by a job.
It helps identify highly specialized positions.

```python
def calculate_advanced_ratio(skills):
    """
    Calculate the ratio of advanced skills.
    """
    if len(skills) == 0:
        return 0
    advanced_count = sum(1 for skill in skills if skill.get("level") == "Advanced")
    return advanced_count / len(skills)

df["advanced_skill_ratio"] = df["skills_normalized"].apply(calculate_advanced_ratio)
```

```python
df["advanced_skill_ratio"].describe()
```

```python
def normalize_skill_name(skill):
    """
    Normalize skill names for matching.
    """
    return (skill.lower().strip().replace(" ", "_"))

def extract_skill_set(skills):
    """
    Convert list of skill dictionaries into normalized skill set.
    """
    return {normalize_skill_name(skill["skill"]) for skill in skills if "skill" in skill}

df["required_skills_set"] = df["skills_normalized"].apply(extract_skill_set)
```

## Experience Level
Experience years are converted into career levels:
- 0 years: Entry Level
- 1-2 years: Junior
- 3-5 years: Mid Level
- 6+ years: Senior
This feature improves job-candidate matching.

```python
df["experience_years"].value_counts().sort_index()
```

```python
def categorize_experience(years):
    """
    Convert numeric experience years into career level.
    """
    if years == 0:
        return "Entry Level"
    elif years <= 2:
        return "Junior"
    elif years <= 5:
        return "Mid Level"
    else:
        return "Senior"
    
df["experience_level"] = df["experience_years"].apply(categorize_experience)
df["experience_level"].value_counts()
```

```python
df["seniority"].value_counts()
```

```python
seniority_mapping = {
    "Worker": 1,
    "Employee": 2,
    "Specialist": 3,
    "Senior Specialist": 4,
    "Manager": 5,
    "Deputy / Senior Manager": 6,
    "CEO / Business Head": 7
}

df["seniority_score"] = df["seniority"].map(seniority_mapping)
```

```python
df["has_salary_info"] = df["salary_min"].notna().astype(int)

df["salary_avg"] = (df["salary_min"] + df["salary_max"]) / 2

df["salary_range"] = (df["salary_max"] - df["salary_min"])
```

```python
df["salary_avg"].quantile([0.25, 0.5, 0.75])
```

```python
salary_q1 = df["salary_avg"].quantile(0.25)
salary_q3 = df["salary_avg"].quantile(0.75)


def categorize_salary(salary):
    """
    Convert salary value into market level.
    """
    if pd.isna(salary):
        return "Unknown"
    elif salary <= salary_q1:
        return "Low"
    elif salary <= salary_q3:
        return "Medium"
    else:
        return "High"

df["salary_level"] = df["salary_avg"].apply(categorize_salary)
df["salary_level"].value_counts()
```
