# Data Glossary
| Column | Description |
|---------|-------------|
| **keyword** | Search keyword used to retrieve the job posting. |
| **job_id** | Unique identifier of the job posting. |
| **title** | Job title. |
| **company** | Employer or company name. |
| **province** | Province where the job is located. |
| **city** | City where the job is located. |
| **salary_visible** | Indicates whether salary information is publicly visible. |
| **experience_years** | Minimum required years of work experience. |
| **description** | Cleaned job description. |
| **work_type** | Employment type (Full Time, Part Time, Internship, etc.). |
| **seniority** | Required seniority level. |
| **category** | Job category assigned by JobVision. |
| **company_size** | Company size based on number of employees. |
| **industries** | Company's industry sectors. |
| **benefits** | Benefits and perks offered by the employer. |
| **salary_title** | Salary range displayed as text. |
| **salary_min** | Minimum salary (Million Tomans). |
| **salary_max** | Maximum salary (Million Tomans). |
| **gender** | Preferred applicant gender, if specified. |
| **required_age_min** | Minimum required age. |
| **required_age_max** | Maximum required age. |
| **military** | Indicates whether military service completion is required. |
| **work_days** | Working schedule. |
| **business_trip** | Required business travel. |
| **software** | Required software, programming languages, and technologies. |
| **software_levels** | Required proficiency level for each listed software skill. |
| **first_activation** | Date when the job was first published. |
| **activation** | Latest activation (refresh) date. |
| **expire** | Job expiration date. |
| **application_count** | Number of submitted applications. |
| **internship** | Indicates whether the position is an internship. |
| **remote** | Indicates whether remote work is available. |
| **urgent** | Indicates whether the vacancy is marked as urgent. |
